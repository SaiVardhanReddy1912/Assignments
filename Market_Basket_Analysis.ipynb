```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("E:\Data Science\Assignments\Association Rules\Association Rules\Online retail.xlsx",header=None)
df.head()
```

```python
# Converting each value as each column
df_new = df[0].str.split(',', expand=True)
df_new
```

```python
trans = []
for i in range(0, 7501):
  trans.append([str(df_new.values[i,j]) for j in range(0, 20)])
```

```python
trans
```

```python
len(trans)
```

```python
# Implementing an Apriori algorithm
from apyori import apriori
rules = apriori(transactions = trans,
                min_support = 0.003,
                min_confidence = 0.2,
                min_lift = 3,
                min_length = 2, max_length = 2)
```

```python
results = list(rules)
results
```

```python
a = []
b = []
c = []
d = []
e = []


for i in range(0,9):
    c.append(results[i][1]) # support
    a.append(results[i][2][0][0]) # base item
    b.append(results[i][2][0][1]) # add item
    d.append(results[i][2][0][2]) # confidence
    e.append(results[i][2][0][3]) # lift
```

```python
d1 = pd.DataFrame(a)
d2 = pd.DataFrame(b)
d3 = pd.DataFrame(c)
d4 = pd.DataFrame(d)
d5 = pd.DataFrame(e)
```

```python
newdata = pd.concat([d1,d2,d3,d4,d5],axis=1)
newdata.columns = ['baseitem','additem','support','cofidence','lift']
newdata
```

```python
# Sort the rules by lift in descending order
sorted_rules = newdata.sort_values('lift', ascending=False)

# Print the top 5 rules with highest lift
print(sorted_rules.head())

# Analyze the rules
# For example, the rule with the highest lift suggests that customers who buy "fromage blanc"
# are very likely to also buy "honey". This could indicate that these items are
# often purchased together.
```

#### 1.What is lift and why is it important in Association rules?

Lift is a measure of how much more likely two items are to be purchased together than if they were purchased independently.
Lift is important in association rules because it helps to identify rules that are truly interesting and actionable. 
Rules with high lift values are more likely to be useful for making recommendations or for understanding customer behavior.

#### 2.	What is support and Confidence. How do you calculate them?

- Support is the frequency of an itemset in the dataset. It is calculated as the number of transactions containing the itemset divided by the total number of transactions.
- Confidence is a measure of the strength of association between two itemsets. It is calculated as the number of transactions containing both itemsets divided by the number of transactions containing the first itemset.
