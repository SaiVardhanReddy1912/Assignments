#### Conversion of Categorical Data into Dummy Variables

```python
data_cat = data.drop(columns=['Volume','Avg Price','Total Sales Value','Discount Rate (%)','Discount Amount','Net Sales Value'])
data_cat.head()
```

```python
data_cat.columns
```

```python
cat_data = data[['Date', 'Day', 'SKU', 'City', 'BU', 'Brand', 'Model']]
cat_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse=False)

# Fit and transform the data
encoded_columns = encoder.fit_transform(cat_data[['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']])

# Create a DataFrame with the encoded columns
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(['Day', 'SKU', 'City', 'BU', 'Brand', 'Model']))

# Combine with the original data
one_hot_encoded_df = cat_data.drop(['Day', 'SKU', 'City', 'BU', 'Brand', 'Model'], axis=1).join(encoded_df)

print(one_hot_encoded_df)
```

```python
## Combining the numerical and enoded categorical dataframes

final_df = pd.concat([data_num_ss,one_hot_encoded_df],axis=1)

final_df.head()
```
